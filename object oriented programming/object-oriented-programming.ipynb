### Static method
we dont use any class method or instance variable. we use @staticmethod decorator. dont need to pass any `self` arguments. 

```python
class Student:
    #by using class name we can access static method
    @staticmethod       #decorator
    def get_personal_details(fname, lname):
        print('your personal details: ', fname, lname)
    
    @staticmethod
    def contact_details(mobile_no, rollno):
        print('your contact details: ', mobile_no, rollno)

Student.get_personal_details('Nidhi', 'Worah')
Student.contact_details(235546708, 1001)
```

## Inheritance
* concept applied for reusability

Types of classes:

- derived class - a class in which properties are inherited are called derived clss or child class

types of inheritance
1. single
2. multilevel
3. multiple

### Single Inheritance
`class derived-class(base-class):`

### Multilevel Inheritance
    class Grandfather:

### Multiple Inheritance

    class Parent1:

```python
#single inheritance
class College: #parent class
    def college_name(self):
        print('College Name : Modern College')
    
class Student(College):     #child class
    def student_info(self):
        print(f'Name : Nidhi Worah \nBranch : AI&ML')

obj = Student()
obj.student_info()
obj.college_name()
```

```python
#multilevel inheritance
class College: #parent class
    def college_name(self):
        print('College Name : Modern College')
    
class Student(College):     #second class - second level
    def student_info(self):
        print(f'Name : Nidhi Worah \nBranch : AI&ML')

class Exam(Student):        #child class
    def subject_info(self):
        print(f'Subject1 : Design Engineering \nSubject2 : Automata Theoy \nSubject3 : Machine Learning')
    
obj = Exam()
obj.student_info()
obj.college_name()
obj.subject_info()
```

```python
#multiple inheritance
class SubjMarks:                #parent class-1
    math = int(input('enter your marks for math: '))
    automata = int(input('enter your marks for automata theory: '))
    python= int(input('enter your marks for python: '))
    english = int(input('enter your marks for english: '))

class PractMarks:               #parent class-2
    python_pracs = int(input('enter your python practical marks: '))

class Result(SubjMarks, PractMarks):            #child class
    def total(self):
        if self.math >= 40 and self.automata >= 40 and self.python >= 40 and self.english >= 40 and self.python_pracs >=20:
            print('pass')
        else:
            print('fail')


obj = Result()
obj.total()
```

## Polymorphism
Ability of a message to be displayed in more than one form. 

2 types :
1. compile time - static binding / early binding. Ex: function overloading
2. run time - dynamic binding / late binding. Ex: function overriding
**overloading**

```python
class Principal:
    def role(self):
        print('Principal - i am managing all the activities in college')

class Dean:
    def role(self):
        print('Dean - i take decisions around here')

class Hod:
    def role(self):
        print('Hod - i am manage teachers and students')

class Faculty:
    def role(self):
        print('Faculty - i teach students')

#-----class declaration completed ------
def func(obj):
    obj.role()  #calling the function

campus = [Principal(), Dean(), Hod(), Faculty()]

for obj in campus:
    func(obj)
```

```python
'''
operator overloading internally implemented using magic method
the __add__ method is a magic method which gets called when we add two numbers using + operator
'''
class Deposit: 
    def __init__(self, cash):
        self.cash = cash
    
    def __add__(self, other):           #magic method
        return self.cash + other.cash

d1 = Deposit(1000)
d2 = Deposit(2000)
print(f'Total cash deposit amount : {d1+d2}')
```

### Method Overriding Concept


```python
class Father:                   #parent class
    def bike(self):
        print('Harley Davidson')
    
    def laptop(self):
        print('Laptop with 2GB RAM and 500 GB HDD I3 processor')

class Aman(Father):
    def laptop(self):
        print('Need a better laptop for programming.')

obj = Aman()
obj.bike()
obj.laptop()            # the laptop function was overwritten by the child class `Aman`
```

Using the __super__() method we can call the function of the parent valss as well. 

```python
class Father:                   #parent class
    def bike(self):
        print('Harley Davidson')
    
    def laptop(self):
        print('Laptop with 2GB RAM and 500 GB HDD I3 processor')

class Aman(Father):
    def laptop(self):
        print('Need a better laptop for programming.')
        super().laptop()                #using the __super__() method

obj = Aman()
obj.bike()
obj.laptop()            # the laptop function was overwritten by the child class `Aman`
```

### Constructor Overriding Method

```python
class Father:
    def __init__(self):
        print('Father - Be at the breakfast table on time!')

class Child(Father):
    def __init__(self):
        print('Child - Ill be late for breakfast')
        super().__init__()              #super method for calling the constructor of the parent class

obj = Child()
```
